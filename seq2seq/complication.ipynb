```python
import pandas as pd
from experiments.data import SOURCE_COLUMN_NAME, TARGET_COLUMN_NAME, filter_ru_adapt, info
from experiments.paths import RU_ADAPT_PATHS, RUSIMPLESENTEVAL_PATH, TEST_DATA_PATH
senteval = pd.read_csv(RUSIMPLESENTEVAL_PATH)
ru_adapt_dfs = []
ru_adapt_dfs.append(filter_ru_adapt(RU_ADAPT_PATHS[0], min_sim = 0.75, min_lev = 20, max_elong_rate = 0.1))
for path in RU_ADAPT_PATHS[1:]:
  ru_adapt_dfs.append(filter_ru_adapt(path, min_sim = 0.5, min_lev = 5, max_elong_rate = 0.1))
info([RUSIMPLESENTEVAL_PATH] + RU_ADAPT_PATHS, [senteval] + ru_adapt_dfs)
```

```python
data = pd.concat([senteval]+ru_adapt_dfs)[[SOURCE_COLUMN_NAME, TARGET_COLUMN_NAME]]
data = data.rename(columns = {SOURCE_COLUMN_NAME: TARGET_COLUMN_NAME, TARGET_COLUMN_NAME: SOURCE_COLUMN_NAME})
data.head()
```

```python
test_data = pd.read_csv(TEST_DATA_PATH)
test_data = test_data.rename(columns = {SOURCE_COLUMN_NAME: TARGET_COLUMN_NAME, TARGET_COLUMN_NAME: SOURCE_COLUMN_NAME})
info([TEST_DATA_PATH], [test_data])
test_data.head()
```

```python
from experiments.data import prepare_data_for_eval
orig, refs = prepare_data_for_eval(data)
print(f'Unique train sents: {len(orig)}')
test_orig, test_refs = prepare_data_for_eval(test_data)
print(f'Unique test sents: {len(test_orig)}')
```

#Training

```python
%cd text_simplifier
```

```python
from seq2seq.paths import get_checkpoints_path
from seq2seq.model import get_model, get_tokenizer
MODEL_ID = 'rut5-base'
DATASET = 'complication'
checkpoints_path = get_checkpoints_path(MODEL_ID, DATASET)
model = get_model(MODEL_ID, checkpoints_path = checkpoints_path)
tokenizer = get_tokenizer(MODEL_ID)
```

```python
from seq2seq.utils import sent_length_info
sent_length_info(data[SOURCE_COLUMN_NAME], tokenizer)
```

```python
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
from seq2seq.utils import RANDOM_STATE, BATCH_SIZE
from seq2seq.dataset import Dataset
print(f'Batch size is {BATCH_SIZE}')
train_data, val_data = train_test_split(data, test_size=0.2, random_state = RANDOM_STATE)
train_dataset = Dataset(train_data, tokenizer, max_length = 50)
print(f'Train data: {len(train_dataset)} sentences')
val_dataset = Dataset(val_data, tokenizer, max_length = 50)
print(f'Val data: {len(val_dataset)} sentences')
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)
```

```python
from experiments.utils import set_logging
logger = set_logging()
%cd text_simplifier
```

```python
from seq2seq.trainer import train
train(model, tokenizer,
      train_loader, val_loader, 
      model_id = MODEL_ID, dataset_name = DATASET, 
      checkpoints_path = checkpoints_path, 
      logger = logger, 
      orig = orig, refs = refs, 
      max_epochs = 20, 
      cleanup_step = 1, report_epoch = 5, save_epoch = 5)
```

```python
text = '14 декабря 1944 года рабочий посёлок Ички был переименован в рабочий посёлок Советский, после чего поселковый совет стал называться Советским.'
input_ids, attention_mask = tokenizer(text, return_tensors = 'pt').values()
input_ids.shape(1)
```
